##### <font color='purple'> R - Square.</font>

R-Squared tells you how much of the variation in the dependent variable (y) is explained by the independent variable(s) (x) in your model.

- **A high R-Squared value (close to 1)** :-  indicates that the model explains most of the variation in the dependent variable.
- **A low R-Squared value (close to 0)** :- indicates that the model explains only a small portion of the variation.

**Example**
- If R-Squared is 0.8, it means that 80% of the variation in the dependent variable is explained by the independent variable(s) in the model.

##### <font color='purple'> RMSE</font>

 RMSE measures the average difference between predicted values and actual values.
- It calculates the difference (error) between each predicted value and actual value.
- Then, it squares these errors, finds the average, and takes the square root.

- **A lower RMSE value indicates** :-  better model performance (more accurate predictions).
- **A higher RMSE value indicates** :- poorer model performance (less accurate predictions).

**Example** :-
If RMSE is 2.5, it means that, on average, your model's predictions are 2.5 units away from the actual values.

##### <font color='purple'> Accuracy Score.</font>

Accuracy Score measures how often your model's predictions are CORRECT.

- It calculates the proportion of correct predictions out of total predictions made.
- Expressed as a percentage (%), ranging from 0% (completely inaccurate) to 100% (perfectly accurate).

- **Higher accuracy score indicates** better performance (more correct predictions).
- **Lower accuracy score indicates** poorer performance (more incorrect predictions).

**Example** :- If Accuracy Score is 85%, it means your model correctly predicted 85% of the instances, and incorrectly predicted 15%.
