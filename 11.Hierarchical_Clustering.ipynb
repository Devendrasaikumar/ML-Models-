# <font color='Blue'>Cluster Analysis  </font>

- Cluster Analysis is a unsupervised learning algorithm. 

- It is a statistical technique used to group similar objects or data points into clusters based on their characteristics.

- The goal is to maximize the similarity within each cluster while minimizing the similarity between different clusters.

# <font color='Blue'>  </font>

# <font color='Blue'>  </font>

### <font color='magenta'>Types of Cluster Analysis  </font>

# <font color='Blue'>  </font>

1. **Hierarchical Clustering**: Builds a tree-like structure of clusters, either agglomerative (bottom-up) or divisive (top-down).

2. **K-Means Clustering**: Partitions data into a predefined number of clusters (k) by minimizing the variance within each cluster.

3. **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**: Groups points that are closely packed together, marking as outliers points that lie alone in low-density regions.

# <font color='Blue'>  </font>

# <font color='red'>Hierarchical Clustering  </font>

- Hierarchical Clustering is a method of cluster analysis that builds an ordered (hierarchy) structure of clusters, starting from individual data points and merging them step by step based on their similarities.

# <font color='Blue'>  </font>

 #### <font color='magenta'> Types of Hierarchical Clustering </font>

*They have two types in this Hierarchical Clustering*

**Agglomerative Clustering (Bottom-Up)**:

- Starts with each data point as its own cluster.


- Iteratively merges the closest pairs of clusters until only one cluster remains or a specified number of clusters is achieved.

**Divisive Clustering (Top-Down)**:

- Starts with one single cluster containing all data points.


- Iteratively splits clusters into smaller ones until each data point is its own cluster or a desired number of clusters is reached.

# <font color='Blue'>  </font>

#### <font color='magenta'>Steps in Agglomerative Clustering   </font>

1.   <font color='green'>**Calculate Distance Matrix** </font>: Compute the pairwise distances (similarities) between data points using metrics like Euclidean distance, Manhattan distance, etc.

*Euclidean distance mathematical formula*

2. <font color='green'>**Linkage Criteria**:</font> Decide how to determine the distance between clusters:

*In this Linkage Criteria they have four types* 

- **Single Linkage** :- Single linkage measures the distance between the closest points of two clusters. It is also known as the nearest neighbor method.

*Example*:

*Clusters*:

Cluster A: {A1 (1, 2), A2 (2, 3)}

Cluster B: {B1 (4, 5), B2 (5, 6)}

--------------------------------------------------------------------------------------------------------------------------------

#### <font color='magenta'>   </font>

- **Complete Linkage** :- Complete linkage measures the distance between the farthest points of two clusters. It is also known as the furthest neighbor method.

*Example*:

*Clusters*:

Cluster A: {A1 (1, 2), A2 (2, 3)}

Cluster B: {B1 (4, 5), B2 (5, 6)}

Calculation: Using previously calculated distances:

Distances:

A1-B1: 4.24

A1-B2: 5.66

A2-B1: 2.83

A2-B2: 4.24

- Farthest distance between A and B is between A1 and B2 (distance of approximately 5.66).

**Result**: The distance between Cluster A and Cluster B is defined as approximately 5.66.

#### <font color='magenta'>   </font>

---------------------------------------------------------------------------------------------------------------------------------

#### <font color='magenta'>   </font>

- **Average Linkage** :- Average linkage computes the average distance between all pairs of points in two clusters.

*Example*:

Clusters:

Cluster A: {A1 (1, 2), A2 (2, 3)}

Cluster B: {B1 (4, 5), B2 (5, 6)}

*Calculation*: Using previously calculated distances:

Distances:

A1-B1: 4.24

A1-B2: 5.66

A2-B1: 2.83

A2-B2: 4.24

**Result**: The distance between Cluster A and Cluster B is defined as approximately 4.25.

#### <font color='magenta'>   </font>

-----------------------------------------------------------------------------------------------------------------------------------

#### <font color='magenta'>   </font>

- **Wards Method** :- Wards method minimizes the total within-cluster variance when merging clusters. This means it seeks to minimize the increase in variance when two clusters are combined.

*Example*:

Clusters:

Cluster A: {A1 (1, 2), A2 (2, 3)}

Cluster B: {B1 (4, 5), B2 (5, 6)}

**Result**: The distance defined by Wards method would result in the most compact clusters after merging, leading to less variance.

#### <font color='magenta'>   </font>

Choosing the right linkage criterion for hierarchical clustering depends on the specific characteristics of your data and the goals of your analysis.

# <font color='red'>Model Building For Divisive Clustering  </font>

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

class DivisiveClustering:
    def __init__(self, data, threshold=0.5):
        self.data = data
        self.threshold = threshold  # Stopping criterion based on variance

    def split_cluster(self, cluster):
        """Split a single cluster into two using K-means if possible."""
        if len(cluster) < 2:  # Ensure there are enough samples to split
            return [cluster]

        kmeans = KMeans(n_clusters=2, random_state=0)
        labels = kmeans.fit_predict(cluster)
        return [cluster[labels == 0], cluster[labels == 1]]

    def recursive_divide(self, cluster):
        """Recursively divide the cluster."""
        if np.var(cluster) < self.threshold or len(cluster) <= 1:
            return [cluster]
        
        split_clusters = self.split_cluster(cluster)
        result = []
        for sub_cluster in split_clusters:
            result.extend(self.recursive_divide(sub_cluster))
        return result

    def fit(self):
        """Fit the divisive clustering model."""
        self.clusters = self.recursive_divide(self.data)
        return self.clusters

# Sample usage
if __name__ == "__main__":
    # Generate sample data
    np.random.seed(42)
    data = np.random.rand(100, 2)  # 100 samples, 2 features

    # Initialize and fit the divisive clustering model
    dc = DivisiveClustering(data, threshold=0.05)
    clusters = dc.fit()

    # Plotting the clusters
    plt.figure(figsize=(10, 6))
    for i, cluster in enumerate(clusters):
        plt.scatter(cluster[:, 0], cluster[:, 1])
        
        # Add labels with some offset
        for point in cluster:
            plt.annotate(f'Cluster {i+1}', 
                         xy=(point[0], point[1]), 
                         xytext=(5, 5),  # Offset the label
                         textcoords='offset points', 
                         fontsize=8)

    plt.title("Divisive Clustering Results")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend()
    plt.show()
```
